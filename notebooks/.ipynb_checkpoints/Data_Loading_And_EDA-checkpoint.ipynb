# Data Cleanup

```python
# Removing Noise text
import re

def clean_text(text):
    # 1. Remove HTML tags (anything between < and >)
    text = re.sub(r'<.*?>', '', text)
    
    # 2. Remove URLs (starts with http or www)
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # 3. Remove Handles (words starting with @)
    text = re.sub(r'@\w+', '', text)
    
    # 4. Remove extra whitespace left behind
    text = " ".join(text.split())
    
    return text
dataset['review'] = dataset['review'].apply(clean_text)
```
